## Create lists & graphs of the best and worst places to swim in the dataset.

"Best" and "Worst" are not the best description and if you look at the data set there is no pass or fail in terms of the amount of EnteroCount found in the testing site. The federal standard for unacceptable water quality is a single sample value of greater than 110 Enterococcus/100mL, or five or more samples with a geometric mean (a weighted average) greater than 30 Enterococcus/100mL. 

To define pass and fail of water quality I am going to look for water samples with less than or equal to 110 as pass and greater than 110 as fail.

#### Is there a relationship between the amount of rain and water quality?

```python
p4 = ggplot(river_keeper, aes(x='FourDayRainTotal', y='EnteroCount')) + geom_point(alpha=.5)
p4 + ylim(0,5000) + xlim(0,10) + stat_smooth(color='blue', method='loess')
```

```python
river_keeper['rainfall_bin'] = (4 * river_keeper['FourDayRainTotal']).round()/4
```

```python
sum_rain = river_keeper.groupby('rainfall_bin').median()
sum_rain['rainfall'] = sum_rain.index
```

```python
sum_rain
```

```python
ggplot(sum_rain,aes(x='rainfall', y='EnteroCount')) + geom_bar(aes(weight='EnteroCount', fill='blue'))
```

If you look at the data you see there is a large spike in your Enterococcus levels in the Hudson River after rainfalls of 2.5, 2.75, and 3.25 inches. There is also a large spike after 8.5 inches, but I think that is an outlier.
